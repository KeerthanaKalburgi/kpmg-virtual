```python
# Define the Agee bins and labels
Agee_bins = [15, 25, 35, 45, 55, 65, float('inf')]
Agee_labels = ['15-24', '25-34', '35-44', '45-54', '55-64','65+']

# Create a new 'Age' column with the Agee categories
df['Age_CI'] = pd.cut(df['Age'], bins=Agee_bins, labels=Agee_labels, right=False)
```
